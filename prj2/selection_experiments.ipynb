### Métodos de seleção
Dois métodos de seleção foram implementados: Roleta e Torneio.

#### Método da roleta
Consiste em dar pesos a cada indivíduo baseado em sua *fitness function* e selecioná-los de forma que o com maiores peso tenham mais chance de serem selecionados. Este método é estocástico, entretanto tende a selecionar os melhores indivíduos, dado que a chance de ser selecionado aumenta de acordo com seu peso. É importante salientar que o método da roleta funciona para problemas de maximização, ou seja, devemos buscar selecionar indivíduos com maior peso. A nossa implementação do problema considera fitness como a quantidade de peças incorretas. Desta forma, o nosso problema é de minimização, portanto, o método da roleta foi desenvolvido utilizando como peso o inverso da fitness para que o problema se torna-se de maximização. Para contornar isto o peso considerado no algoritmo foi:

**1 / (fitness + 1)**

Como visto na fórmula acima, utilizando o inverso da *fitness + 1*, para evitar divisão por zero, uma vez que quando não há peças incorretas, a fitness será zero.

#### Método do torneio
Consisteem dois passos: a seleção dos participantes do torneio e a competição do torneio:
- A **seleção dos participantes** é feita de forma estocástica e **qualquer indivíduo da população possui a mesma chance de ser selecionado** para participar do torneio. O número de participantes selecionados (tamanho do torneio) é definida na implementação. É importante notar que quanto maior o tamanho do torneio maior a chance de convergir prematuramente. Como visto no relatório, ao utilizar um tamanho de torneio de 10% da população total, não atingimos uma solução correta, enquanto ao utilizar 1% da população total atingimos a solução correta.
- A **competição** consiste em percorrer os participantes e selecionar o participante com maior fitness. Note que, o método de torneio é bastante versátil, se considerarmos que ele funciona bem para problemas de maximização e minimização da mesma forma. Ainda, ele pode ser utilizado com valores negativos, enquanto o método da roleta não funciona bem com número negativos, dado que ele é baseado na probabilidade de que um indivíduo seja selecionado relativo a sua *fitness* (sendo que *fitness* negativas podem levar a probabilidades negativas, o que não faz sentido matemático).

Este experimento demonstra que **um porcento** da população total é uma boa candidata a ordem de grandeza para tamanho do torneio.
É interessante considerar que **quanto maior o tamanho do torneio, maior a tendência de selecionar vários indivíduos repetidos**, uma vez que o vencedor do torneio será sempre o indívduo com menor fitness. Imagine que temos uma população de 1000 indivíduos e um tamanho de torneio de 1000. Sempre escolheremos a população inteira para o torneio e consequentemente, o indivíduo selecionado será sempre o de menor fitness. Logo, considerando uma população em que todos os indivíduos tem fitness diferentes, o resultado de nossa seleção será 1000 indivíduos repetidos.
Ainda, **quanto maior o tamanho do torneio, maior o processamento necessário**, tanto para escolher participantes, quanto para a competição. Imagine que temos um tamanho do torneio de 100 indivíduos. Para escolher os participantes teremos 100 iterações e para a competição executaremos mais 100 iterações. Se quisermos selecionar 100 indivíduos no total, teremos ***100 * (100 + 100) = 20000 iterações** para selecionar todos os indivíduos.
